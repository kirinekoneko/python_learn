```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neural_network import MLPRegressor

sample_size = 100

X = np.linspace(0, 6, sample_size)
y = np.sin(X)

for i in range(100):
    size = (i + 1) * 10
    hidden_layer_sizes = (size, )
    model = MLPRegressor(hidden_layer_sizes=hidden_layer_sizes)
    model.fit(X.reshape((sample_size, 1)), y)
    y_pred = model.predict(X.reshape((sample_size, 1)))

    plt.figure()
    plt.scatter(X, y)
    plt.scatter(X, y_pred)
    plt.title(str(hidden_layer_sizes))
```

```python
import numpy as np

sample_size = 1000
X = np.random.uniform(-10, 10, (sample_size, 2))

# plt.scatter(X[:, 0], X[:, 1])


radius = 8

y = np.zeros(sample_size)

for i in range(sample_size):
    if np.sqrt(X[i, 0] ** 2 ++ X[i, 1] ** 2) <= radius:
        y[i] = 1
    else:
        y[i] = 0
        

# xy11
plt.figure(figsize=(10, 10))
plt.scatter(X[y == 0, 0], X[y == 0, 1])
plt.scatter(X[y == 1, 0], X[y == 1, 1])
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(X, y)

X_test = np.random.uniform(-10, 10, (10000, 2))
y_pred = model.predict(X_test)

plt.figure(figsize=(5,5))
plt.scatter(X_test[y_pred == 0, 0], X_test[y_pred == 0, 1])
plt.scatter(X_test[y_pred == 1, 0], X_test[y_pred == 1, 1])
```

```python
from sklearn.neighbors import KNeighborsClassifier

model = KNeighborsClassifier()
model.fit(X, y)

y_pred = model.predict(X_test)

plt.figure(figsize=(5,5))
plt.scatter(X_test[y_pred == 0, 0], X_test[y_pred == 0, 1])
plt.scatter(X_test[y_pred == 1, 0], X_test[y_pred == 1, 1])
```

```python
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier(criterion='entropy')
# model = DecisionTreeClassifier()
model.fit(X, y)

y_pred = model.predict(X_test)

plt.figure(figsize=(5,5))
plt.scatter(X_test[y_pred == 0, 0], X_test[y_pred == 0, 1])
plt.scatter(X_test[y_pred == 1, 0], X_test[y_pred == 1, 1])
```

```python
from sklearn.neural_network import MLPClassifier


model = MLPClassifier(hidden_layer_sizes=(1000, ))
model.fit(X, y)


y_pred = model.predict(X_test)

plt.figure(figsize=(5,5))
plt.scatter(X_test[y_pred == 0, 0], X_test[y_pred == 0, 1])
plt.scatter(X_test[y_pred == 1, 0], X_test[y_pred == 1, 1])
```
